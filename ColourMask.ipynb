```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reference RGB values for the colours in the image
RefVals = [np.array([20,20,20]) ,np.array([130,0,220]),np.array([51,119,255]) ,np.array([2,208,161]) ,np.array([160,230,50]) ,np.array([230,220,50]),np.array([230,175,45]) ,np.array([240,130,40]) ,np.array([250,60,60]) ,np.array([255,128,192]) ,np.array([255,180,220]) ]
```

```python
from os import listdir
from os.path import isfile, join
#Read the each image file and put them to an list
mypath='sea/'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = np.empty(len(onlyfiles), dtype=object)

for n in range(0, len(onlyfiles)):
  images[n] = cv2.imread( join(mypath,onlyfiles[n]) )
  
```

```python
resized_image= np.empty(len(onlyfiles), dtype=object)
crop_img= np.empty(len(onlyfiles), dtype=object)
RGB_img= np.empty(len(onlyfiles), dtype=object)
#Iterate through all the images
for n in range(0, len(images)):
    #Crop the image
    crop_img[n]=images[n][1930:2318,795:984]
    #Resize the image
    newsize = (230, 388)
    resized_image[n]= cv2.resize(crop_img[n], newsize) 
    #Convert the image into RGB scale
    RGB_img[n] = cv2.cvtColor(resized_image[n], cv2.COLOR_BGR2RGB)
    plt.imshow(RGB_img[n])
    plt.show()
    
    #Initialized the normalized mask
    msk_norm_tot=0
    rr = 0
    plt.figure(figsize=(10, 10))
    #Iterate through the reference colours
    while rr <=10:
        #Identifying the colour wise mask
        msk = cv2.inRange(RGB_img[n], RefVals[rr]-20, RefVals[rr]+20)
        #Get the normalized mask
        msk_norm = msk/255
        print(msk_norm)
        #Get the sum of the normalized mask
        msk_norm_tot=msk_norm_tot+msk_norm
        #Get the resulting image segments within the mask
        result = cv2.bitwise_and(RGB_img[n], RGB_img[n], mask=msk)
        plt.subplot(4, 3, rr+1)  
        plt.imshow(result)
        rr = rr +1
    
    plt.show()
    #Convert the binary mask to the range 0 to 255
    result_mask=255*(msk_norm_tot)
    #Clip the mask between 0 and 255 and convert the mask to the range of 0 to 1
    result_mask=result_mask.clip(0,255).astype("uint8")
    print(result_mask)
    #Get the normalized mask
    result_mask_norm = result_mask/255
    print(result_mask_norm)
    #Applying the obtained mask to the original image
    result_image= cv2.bitwise_and(RGB_img[n], RGB_img[n], mask=result_mask)
    plt.imshow(result_image)
    plt.show()
    
    
```
